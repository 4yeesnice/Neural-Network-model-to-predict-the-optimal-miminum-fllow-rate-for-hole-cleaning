```python
def prepare_data_docx(filename):
    doc = Document(filename)

    all_tables = doc.tables
#     print('Всего таблиц в документе:', len(all_tables))
#     print(filename)
    # создаем пустой словарь под данные таблиц
    data_tables = {i:None for i in range(len(all_tables))}
    # проходимся по таблицам
    for i, table in enumerate(all_tables):
    #     print('\nДанные таблицы №', i)
        # создаем список строк для таблицы `i` (пока пустые)
        data_tables[i] = [[] for _ in range(len(table.rows))]
        # проходимся по строкам таблицы `i`
        for j, row in enumerate(table.rows):
            # проходимся по ячейкам таблицы `i` и строки `j`
            for cell in row.cells:
                # добавляем значение ячейки в соответствующий
                # список, созданного словаря под данные таблиц
                data_tables[i][j].append(cell.text)





    height_of_collars = []
    diameter_of_collars = []


    for i in range(3):
        try:
            if 'Open Hole' in data_tables[4][i]:
                Depth_Open = float(data_tables[4][i][1])
                ID_Open = float(data_tables[4][i][4])

            elif 'Casing' in data_tables[4][i]:
                Depth_Cas = float(data_tables[4][i][1])
                ID_Cas = float(data_tables[4][i][4])
                
        except ValueError:
            ID_Open = 152.40


#     print(Depth_Open)            
#     print(ID_Open)            
#     print(Depth_Cas)            
#     print(ID_Cas)

    # Depth_Cas = float(data_tables[4][1][1])
    # ID_Cas = float(data_tables[4][1][4])
    # ID_Open = float(data_tables[4][2][4])


    constraint = len(data_tables[5])

    for i in range(2, constraint):
        if "," in data_tables[5][i][2]:
            height_of_collars.append(float(data_tables[5][i][2].replace(",", "")))
            diameter_of_collars.append(float(data_tables[5][i][3]))
        else:
            height_of_collars.append(float(data_tables[5][i][2]))
            diameter_of_collars.append(float(data_tables[5][i][3]))

    old_height = height_of_collars[:]        
    old_diameter = diameter_of_collars[:]

    minimum = old_height.index(min(old_height))
    height_of_collars = old_height[minimum:]
    height_of_collars = height_of_collars + old_height[:minimum]

    diameter_of_collars = old_diameter[minimum:]
    diameter_of_collars = diameter_of_collars + old_diameter[:minimum]


    # print(height_of_collars)
    # print(diameter_of_collars)

    height = data["Расстояние вдоль колонны"]




    def calculate_capacity(inner_diameter, outer_diameter):
    #     print(inner_diameter, 'inn')
    #     print(outer_diameter, 'out')
        capacity = (inner_diameter**2 - outer_diameter**2) / 1273240
        return capacity

    def check_diameter_for_height(height):
        for depth in height_of_collars:
    #         print(height, depth)
            if height >= depth and height <= height_of_collars[height_of_collars.index(depth)+1]:
                index = height_of_collars.index(depth)
    #             print(index, 'index')
                return diameter_of_collars[index]
            elif height < depth:
                return diameter_of_collars[0]
            elif height > max(height_of_collars):
                return diameter_of_collars[-1]

    capacity = []        
    open_hole = []
    def check_capacity_for_height(glubina):
        if glubina >= Depth_Cas:
            outer = check_diameter_for_height(glubina)
            capac = calculate_capacity(ID_Open, outer)
            capacity.append(capac)
            open_hole.append(1)
        else:
            outer = check_diameter_for_height(glubina)
            capac = calculate_capacity(ID_Cas, outer)
            capacity.append(capac)
            open_hole.append(0)

    for glubina in height:
        check_capacity_for_height(glubina)







    linear_capacity_casing = float(data_tables[4][1][8])
    linear_capacity_openhole = float(data_tables[4][2][8])

    linear_capacity = []

    def check_linear_capacity_for_height(height):
        if height >= Depth_Cas:
            linear_capacity.append(linear_capacity_openhole)
        else:
            linear_capacity.append(linear_capacity_casing)

    for glubina in height:
         check_linear_capacity_for_height(glubina)




    friction_casing = float(data_tables[4][1][7])
    friction_openhole = float(data_tables[4][2][7])

    friction = []


    def check_friction_for_height(height):
        if height >= Depth_Cas:
            friction.append(friction_openhole)
        else:
            friction.append(friction_casing)

    for glubina in height:
         check_friction_for_height(glubina)



    for i in range(len(data_tables)):
        for j in range(len(data_tables[i])):
            if 'Механическая скорость проходки' in data_tables[i][j]:
                ROP = data_tables[i][j][1]
                RPM = data_tables[i][j][3]
                Cut_diam = data_tables[i][j+1][1]
                Cut_density = data_tables[i][j+1][3]
                Cut_porosity = data_tables[i][j+2][1]
            elif 'Подача насосов' in data_tables[i][j]:
                Pump = data_tables[i][j][1]

            elif 'Учет температуры' in data_tables[i][j]:
                Fluid_density = data_tables[i][j+1][2]
                plastic_viscosity = data_tables[i][j+1][4]
                yield_strength = data_tables[i][j+1][5]
                
            elif 'Начальная глубина по стволу' in data_tables[i][j]:
                Pressure_stand = data_tables[i+1][0][3]
                Pressure_loss = data_tables[i+1][1][1]
                Bit_flow = data_tables[i+1][2][3]
                Hyd_pow = data_tables[i+1][3][1]


    # ROP = data_tables[22][0][1]
    # RPM = data_tables[22][0][3]
    # Cut_diam = data_tables[22][1][1]
    # Cut_density = data_tables[22][1][3]
    # Pump = data_tables[26][0][1]
    # Cut_porosity = data_tables[22][2][1]

    ROP = list(map(float, re.findall(r'\d.\d{2}', ROP)))*len(friction)
    RPM = list(map(float, re.findall(r'\d{2}', RPM)))*len(friction)
    Pump = list(map(float, re.findall(r'\d{2}.\d{3,}', Pump)))*len(friction)
    Cut_porosity = list(map(float, re.findall(r'\d{2}.\d', Cut_porosity)))*len(friction)
    Cut_diam = list(map(float, re.findall(r'\d.\d{2}', Cut_diam )))*len(friction)
    Cut_density= list(map(float, re.findall(r'\d.\d{2}', Cut_density)))*len(friction)
    Pressure_stand = list(map(float, re.findall(r'\d{1,}.\d{1,}', Pressure_stand)))*len(friction)
    Pressure_loss = list(map(float, re.findall(r'\d{1,}.\d{1,}', Pressure_loss)))*len(friction)
    Bit_flow = list(map(float, re.findall(r'\d{1,}.\d{1,}', Bit_flow)))*len(friction)
    Hyd_pow = list(map(float, re.findall(r'\d{1,}.\d{1,}', Hyd_pow)))*len(friction)
    
    try:
        Fluid_density = [int(Fluid_density) for i in range(len(friction))]
        plastic_viscosity = [int(plastic_viscosity) for i in range(len(friction))]
        yield_strength = [int(yield_strength) for i in range(len(friction))]
    except ValueError:
        Fluid_density = [1125 for i in range(len(friction))]
        plastic_viscosity = [16 for i in range(len(friction))]
        yield_strength = [25 for i in range(len(friction))]


    def percents(number):
        return number / 100

    Cut_porosity = list(map(percents, Cut_porosity))
    data.insert(3, "Capacity", capacity) 
    data.insert(3, "Linear Capacity", linear_capacity)
#     data.insert(3, "Friction", friction)  
#     data.insert(3, "Cut_diam", Cut_diam)
#     data.insert(3, "Cut_density", Cut_density)
    data.insert(3, "Open Hole", open_hole)
    data.insert(3, "Pressue at stand", Pressure_stand)
    data.insert(3, "Pressure loss", Pressure_loss)
    data.insert(3, "Bit flow", Bit_flow)
    data.insert(3, "Hydraulic power", Hyd_pow)
#     data.insert(3, "ROP", ROP)
#     data.insert(3, "RPM", RPM)
    data.insert(3, "Pump volume", Pump)
#     data.insert(3, "Cuttings Porosity", Cut_porosity)

#     data.insert(3, "Fluid_density", Fluid_density)
#     data.insert(3, "plastic_viscosity", plastic_viscosity)
#     data.insert(3, "yield_strength", yield_strength)
```
